```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
fifa_data = pd.read_csv('fifa_data.csv')
```

```python
df = pd.read_csv('fifa_data.csv')
df.head(5)
```

# Question 1

```python
top_countries = fifa_data['Nationality'].value_counts()
print(f"Country with most players: {top_countries.idxmax()} ({top_countries.max()} players)")
```

# Question 2

```python
top_5_countries = top_countries.head(5)
plt.figure(figsize=(10, 6))
top_5_countries.plot(kind='bar', color='skyblue')
plt.title('Top 5 Countries with Most Players')
plt.xlabel('Country')
plt.ylabel('Number of Players')
plt.show()
```

# Question 3

```python
fifa_data['Wage'] = fifa_data['Wage'].replace({'€': '', 'K': '000', 'M': '000000'}, regex=True).astype(float)
highest_paid_player = fifa_data.loc[fifa_data['Wage'].idxmax(), ['Name', 'Wage']]
print(f"Highest paid player: {highest_paid_player['Name']} with a salary of €{highest_paid_player['Wage']}")
```

# Question 4

```python
plt.figure(figsize=(10, 6))
plt.hist(fifa_data['Wage'].dropna(), bins=30, color='lightgreen', edgecolor='black')
plt.title('Salary Range of FIFA Players')
plt.xlabel('Salary (in Euros)')
plt.ylabel('Number of Players')
plt.show()
```

# Question 5

```python
fifa_data['Height'] = fifa_data['Height'].str.replace("'", ".").astype(float)
tallest_player = fifa_data.loc[fifa_data['Height'].idxmax(), ['Name', 'Height']]
print(f"Tallest player: {tallest_player['Name']} with a height of {tallest_player['Height']} feet")
```

# Question 6

```python
top_club = fifa_data['Club'].value_counts().idxmax()
print(f"Club with the most players: {top_club}")
```

# Question 7

```python
preferred_foot = fifa_data['Preferred Foot'].value_counts()
print(f"Most preferred foot: {preferred_foot.idxmax()} with {preferred_foot.max()} players")
plt.figure(figsize=(6, 4))
preferred_foot.plot(kind='bar', color='orange')
plt.title('Preferred Foot of FIFA Players')
plt.xlabel('Foot')
plt.ylabel('Number of Players')
plt.show()
```
